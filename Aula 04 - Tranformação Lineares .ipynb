Tranformação Linear

Um transformação linear é uma função entre dois espaços vetoriais que preserva as operações de adição de vetores e de multiplicação de um vetor por um escalar. Em outras palavras, seja V e W dois espaços vetoriais, uma função T: V -> W é uma transformação linear se possuir as seguintes propriedades:

* T(u + v) = T(u) + T(v), para todo u e v em V.
* T(ku) = kT(u), para todo u em V e todo escalar k.

Essas propriedades podem ser resumidas em uma única propriedade:

T(u + kv) = T(u) + kT(v), para todo u e v em V e todo escalar k.

No caso especial em que V = W, a transformação linear é chamada um operador linear de V.

Representando vetore.


```python
import matplotlib.pyplot as plt


# Definindo o vetor
vetor = [2, 2]


# Definindo as origens para o vetor
origens = [(0, -1), (0, 1), (0, 0), (0, 2)]

# Extraindo as coordenadas x e y das origens
origens_x = [origem[0] for origem in origens]
origens_y = [origem[1] for origem in origens]

 # Plotando o vetor (2, 2) com as origens especificadas
plt.figure()
plt.quiver(origens_x, origens_y, [vetor[0]]*len(origens_x), [vetor[1]]*len(origens_y), angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(-1, 5)
plt.ylim(-2, 5)

 # Configurações adicionais do gráfico

plt.xlim(-1, 5)
plt.ylim(-2, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Um vetor com diferentes potnro de origem')

# Exibindo gráfico
plt.show()
```

Base canônica
A base canônioca, frequentemente referida como base padrão, é uma coleção de vetores são usados como base para

Para um espaço tridimensional, a base canônica é composta por três vatores:

* i: Vetor unitário ao longo do eixo x(1, 0, 0).
* j: Vetor unitário ao longo do eixo y(0, 1, 0).
* k: Vetor unitário ao longo do eixo z(0, 0, 1).

Esses vetores formam uma base para o espaço tridimensional, o que significa que qualquer vetor neste espaço pode ser representado como uma combinação linear desses vetores.

Para um espaço bidimensional, a base canônica é composta por dois vetores:

* i: Vetor unitário ao longo do eixo x(1, 0).
* j: Vetor unitário ao longo do eixo y(0, 1).

```python
import matplotlib.pyplot as plt

# Definindo o vetor na base canônica
vetor_x = [1, 0]
vetor_y = [0, 1]

# Plotando o vetor na base canônica
plt.figure()
plt.quiver(0, 0, vetor_x[0], vetor_x[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, vetor_y[0], vetor_y[1], angles='xy', scale_units='xy', scale=1, color='r')

# Configurações adicionais do gráfico
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Vetor na base canônica')

# Exibindo gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definindo o vetor na base canônica
vetor_x = [1, 0] # Vetor unitário ao longo do eixo x
vetor_y = [0, 1] # Vetor unitário ao lonfo do eixo y

# Plotando o vetor na base canônica
plt.figure()
plt.quiver(0, 0, vetor_x[0], vetor_x[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, vetor_y[0], vetor_y[1], angles='xy', scale_units='xy', scale=1, color='r')

# Adicionando o vetor(4, 3)
plt.quiver(0, 0, 4, 3, angles='xy', scale_units='xy', scale=1, color='g')

# Configurações adicionais do gráfico
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Combinações feitas com os vetores da base canônica')

# Exibindo gráfico
plt.show()
```
