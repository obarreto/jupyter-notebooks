```python
import numpy as np
import matplotlib.pyplot as plt

def translacao_2d(ponto, deslocamento):
  #Passo 1: Converter o ponto para coordenadas homogêneas
  ponto_homogeneo = np.array([ponto[0], ponto[1], 1])

  #Passo 2: Criar a matriz de translação
  matriz_translacao = np.array([[1, 0, deslocamento[0]], #dx é o deslocamento na direção X
                                [0, 1, deslocamento[1]], #dy é o deslocamento na direção y
                                [0, 0, 1]]) # A última linha é fixa para coordenadas homegêneas

  #Passo 3: Aplicar a translação multiplicando a matriz pelo ponto
  novo_ponto_homogeneo = np.dot(matriz_translacao, ponto_homogeneo)

  #Passo 4: Converter de volta para coordenadas 2D
  novo_ponto = novo_ponto_homogeneo[:2]

  return novo_ponto

#Exemplo de uso
ponto_inicial = [2, 3]
deslocamento = [4, 5]
novo_ponto = translacao_2d(ponto_inicial, deslocamento)
print("Ponto inicial:", ponto_inicial)
print("Deslocamento:", deslocamento)
print("Novo ponto:", novo_ponto)

# Suponha que ponto = (2, 3) e novo_ponto = (5, 4)
plt.plot(ponto_inicial[0], ponto_inicial[1], 'ro', label='Ponto Original')
plt.plot(novo_ponto[0], novo_ponto[1], 'bo', label='Novo Ponto')

# Adicione rótulos e legenda
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Translação de Ponto')
plt.grid(True)

# Exiba o gráfico
plt.show()
```
