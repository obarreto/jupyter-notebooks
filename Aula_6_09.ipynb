```python
import numpy as np

# Define as coordenadas dos pontos
ponto_x = np.array([100, 30])  # Coordenadas (x, y) do ponto X
ponto_y = np.array([30, 40])  # Coordenadas (x, y) do ponto Y

# Calcula a diferença entre as coordenadas dos pontos
diferenca = ponto_y - ponto_x

# Calcula a distância euclidiana usando a norma L2
distancia = np.linalg.norm(diferenca)

print(f"A distância entre os pontos X e Y é: {distancia:.3f}")
```

```python
import matplotlib.pyplot as plt # Importa a biblioteca Matplotlib para plotar gráficos
import numpy as np # Importa a biblioteca NumPy para manipulação de arrays

# Solicita as coordenadas dos pontos ao usuário
x1 = int(input("Digite a coordenada x do primeiro ponto: ")) # Recebe a coordenada x do primeiro ponto e converte para inteiro
y1 = int(input("Digite a coordenada y do primeiro ponto: ")) # Recebe a coordenada y do primeiro ponto e converte para inteiro
x2 = int(input("Digite a coordenada x do segundo ponto: ")) # Recebe a coordenada x do segundo ponto e converte para inteiro
y2 = int(input("Digite a coordenada y do segundo ponto: ")) # Recebe a coordenada y do segundo ponto e converte para inteiro

# Cria uma imagem em branco
imagem = np.zeros((100, 100, 3), dtype=np.uint8) # Cria um array NumPy de zeros com dimensões 100x100x3 (altura, largura, canais de cor) e tipo de dados uint8 (inteiro sem sinal de 8 bits)

# Desenha os pontos na imagem
imagem[y1, x1] = [255, 0, 0]  # Define o pixel na posição (y1, x1) como vermelho (RGB: 255, 0, 0)
imagem[y2, x2] = [0, 0, 255]  # Define o pixel na posição (y2, x2) como azul (RGB: 0, 0, 255)

# Calcula a distância entre os pontos
distancia = np.sqrt((x2 - x1)**2 + (y2 - y1)**2) # Calcula a distância euclidiana usando o teorema de Pitágoras: raiz quadrada da soma dos quadrados das diferenças das coordenadas

# Desenha uma linha entre os pontos
plt.imshow(imagem) # Exibe a imagem criada
plt.plot([x1, x2], [y1, y2], color='green', linewidth=3) # Desenha uma linha verde entre os pontos (x1, y1) e (x2, y2) com largura 2

# Exibe a distância na imagem
plt.text((x1 + x2) / 2, (y1 + y2) / 2, f"{distancia:.2f}", color='green', fontsize=24, ha='center', va='center') # Adiciona um texto com a distância calculada no ponto médio da linha

# Configura o título e os eixos
plt.title("Distância entre dois pontos") # Define o título do gráfico
plt.xlabel("Eixo X") # Define o rótulo do eixo x
plt.ylabel("Eixo Y") # Define o rótulo do eixo y

# Exibe a imagem
plt.show() # Mostra o gráfico com a imagem, os pontos, a linha e a distância
```

```python
import matplotlib.pyplot as plt # Importa a biblioteca Matplotlib para plotar gráficos
import numpy as np # Importa a biblioteca NumPy para manipulação de arrays

# Coordenadas dos pontos
ponto1 = np.array([1, 2, 3])  # Define as coordenadas do primeiro ponto como um array NumPy
ponto2 = np.array([4, 5, 6])  # Define as coordenadas do segundo ponto como um array NumPy

# Função para calcular a diferença entre pontos
def calcular_diferenca(p1, p2): # Define uma função chamada calcular_diferenca que recebe dois pontos como entrada
  return p2 - p1 # Retorna a diferença entre as coordenadas dos pontos

# Retangulares
diferenca_retangular = calcular_diferenca(ponto1, ponto2) # Chama a função calcular_diferenca para calcular a diferença entre os pontos em coordenadas retangulares
print(f'Diferença (Retangulares): {diferenca_retangular}') # Imprime a diferença calculada

# Polares (apenas 2D)
ponto1_polar = np.array([np.sqrt(ponto1[0]**2 + ponto1[1]**2), np.arctan2(ponto1[1], ponto1[0])]) # Converte as coordenadas do primeiro ponto para polares (raio e ângulo)
ponto2_polar = np.array([np.sqrt(ponto2[0]**2 + ponto2[1]**2), np.arctan2(ponto2[1], ponto2[0])]) # Converte as coordenadas do segundo ponto para polares (raio e ângulo)
diferenca_polar = calcular_diferenca(ponto1_polar, ponto2_polar) # Chama a função calcular_diferenca para calcular a diferença entre os pontos em coordenadas polares
print(f'Diferença (Polares): {diferenca_polar}') # Imprime a diferença calculada

# Cilíndricas
ponto1_cilindrico = np.array([np.sqrt(ponto1[0]**2 + ponto1[1]**2), np.arctan2(ponto1[1], ponto1[0]), ponto1[2]]) # Converte as coordenadas do primeiro ponto para cilíndricas (raio, ângulo, altura)
ponto2_cilindrico = np.array([np.sqrt(ponto2[0]**2 + ponto2[1]**2), np.arctan2(ponto2[1], ponto2[0]), ponto2[2]]) # Converte as coordenadas do segundo ponto para cilíndricas (raio, ângulo, altura)
diferenca_cilindrica = calcular_diferenca(ponto1_cilindrico, ponto2_cilindrico) # Chama a função calcular_diferenca para calcular a diferença entre os pontos em coordenadas cilíndricas
print(f'Diferença (Cilíndricas): {diferenca_cilindrica}') # Imprime a diferença calculada

# Esféricas
ponto1_esferico = np.array([np.sqrt(ponto1[0]**2 + ponto1[1]**2 + ponto1[2]**2), np.arctan2(np.sqrt(ponto1[0]**2 + ponto1[1]**2), ponto1[2]), np.arctan2(ponto1[1], ponto1[0])]) # Converte as coordenadas do primeiro ponto para esféricas (raio, ângulo polar, ângulo azimutal)
ponto2_esferico = np.array([np.sqrt(ponto2[0]**2 + ponto2[1]**2 + ponto2[2]**2), np.arctan2(np.sqrt(ponto2[0]**2 + ponto2[1]**2), ponto2[2]), np.arctan2(ponto2[1], ponto2[0])]) # Converte as coordenadas do segundo ponto para esféricas (raio, ângulo polar, ângulo azimutal)
diferenca_esferica = calcular_diferenca(ponto1_esferico, ponto2_esferico) # Chama a função calcular_diferenca para calcular a diferença entre os pontos em coordenadas esféricas
print(f'Diferença (Esféricas): {diferenca_esferica}') # Imprime a diferença calculada

# Plotagem dos pontos (apenas para coordenadas retangulares)
fig = plt.figure() # Cria uma nova figura para o gráfico
ax = fig.add_subplot(111, projection='3d') # Adiciona um subplot 3D à figura (111 significa 1 linha, 1 coluna, 1º subplot)

# Plota os pontos
ax.scatter(ponto1[0], ponto1[1], ponto1[2], c='r', marker='o') # Plota o ponto1 com a cor vermelha ('r') e marcador 'o'
ax.scatter(ponto2[0], ponto2[1], ponto2[2], c='b', marker='o') # Plota o ponto2 com a cor azul ('b') e marcador 'o'

# Configura os eixos
ax.set_xlabel('X') # Define o rótulo do eixo X
ax.set_ylabel('Y') # Define o rótulo do eixo Y
ax.set_zlabel('Z') # Define o rótulo do eixo Z

plt.show() # Exibe o gráfico
```

```python
import numpy as np

def translacao_2d(ponto, deslocamento):
    #Passo 1: Converter o ponto para coordenadas homogêneas
    ponto_homogeneo = np.array(ponto[0], ponto[1], 1)
    # Em cordenadas homogêas, um ponto 2d(x,y) é representado
    #como (x, y, 1). Adicionar 1 no fia
    
    #Passo 2: Criar a matriz de translação
    matriz_translacao = np.array([[1, 0, deslocamento[0]], #dx é o deslocamento na direção x
                                  [0, 1, deslocamento[1]],  #dy é o deslocamento na direção y
                                  [0, 0, 1]])# A última linha é fixa para coordenadas homogêneas
    
    #Passo 3: Aplicar a translação multiplicando a matriz pelo ponto
    novo_ponto_homogeneo = np.dot(matriz_translacao, ponto_homogeneo)
    # .dot é uma função de bibliotecas numpy para fazer multiplicação de matrizes
    
    #Passo 4: Converter de volta para coordenadas 2D
    novo_ponto = novo_ponto_homogeneo[:2]
    return novo_ponto
```

```python
#Exemplo de uso
ponto_inicial = [2, 3]
deslocamento = [4, 5]
novo_ponto = translacao_2d(ponto_incial, deslocamento)
print("Ponto Incicial", ponto_incial)
print("Deslocamento:", deslocamento)
print("Novo ponto", novo_ponto)

# Exiba o gráfico
plt.show()
```
