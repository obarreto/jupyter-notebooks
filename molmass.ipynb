```python
from molmass import Formula
```

# ETANOL

```python
formula = Formula ('C2H5OH')
formula.mass #g/mol
```

```python
##O Molmass aceita diferentes entrada para um mesmo composto. Vamos passar outra
## string com abreviao do radical etil na orgnica, Et, para obter a massa molar do etanol:

formula = Formula('EtOH')
formula.mass
```

#### Para obter informaes da composio elementar do composto. podemos utilizar o mtodo 'composition'

```python
formula.composition()
```
