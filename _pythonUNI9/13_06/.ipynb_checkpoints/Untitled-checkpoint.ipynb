```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
vendas = pd.read_csv('VENDAS.csv')
```

```python
vendas
```

```python
vendas.info()
```

```python
#removendo os R e $ com replace
vendas['Preço']= vendas['Preço'].str.replace('R','',regex=True)
vendas['Preço']= vendas['Preço'].str.replace('$','',regex=True)
vendas['Custo']= vendas['Custo'].str.replace('R','',regex=True)
vendas['Custo']= vendas['Custo'].str.replace('$','',regex=True)
vendas
```

```python
vendas.dtypes
```

```python
#vendas['Preço']= vendas['Preço'].str.replace(',','.', regex=True).astype(float)
vendas['Custo']= vendas['Custo'].str.replace(',','.', regex=True).astype(float)
```

```python
vendas.fillna(vendas['Preço'].mean())
```
