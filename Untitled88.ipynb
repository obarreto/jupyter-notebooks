```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV
file_path = "/media/obarreto/MSDOSX/_science-computer/jupyter-notebooks/SampleSearPointP2VanDiemen2017.csv"
df = pd.read_csv(file_path, skiprows=14, delimiter=",")  # Mantendo o delimitador como ponto e vírgula

# Exibir as colunas disponíveis para garantir que estamos usando os nomes corretos
print(df.columns)

# Verificar se as colunas 'GPS Latitude' e 'GPS Longitude' existem
if 'GPS Latitude' in df.columns and 'GPS Longitude' in df.columns:
    # Converter para float, forçando erros a se tornarem NaN
    df['GPS Latitude'] = pd.to_numeric(df['GPS Latitude'], errors='coerce')
    df['GPS Longitude'] = pd.to_numeric(df['GPS Longitude'], errors='coerce')

    # Remover linhas com valores NaN em latitude ou longitude
    df = df.dropna(subset=['GPS Latitude', 'GPS Longitude'])

    # Usar 'GPS Latitude' e 'GPS Longitude' para plotar o traçado da pista
    latitude = df['GPS Latitude']
    longitude = df['GPS Longitude']

    # Plotar o traçado da pista
    plt.figure(figsize=(10, 6))
    plt.plot(longitude, latitude, marker='o', linestyle='-', color='b')
    plt.title("Traçado da Pista")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True)
    plt.axis('equal')
    plt.show()
else:
    print("As colunas 'GPS Latitude' e 'GPS Longitude' não foram encontradas.")
```

```python
import pandas as pd

# Carregar os dados do CSV
file_path = "/media/obarreto/MSDOSX/_science-computer/jupyter-notebooks/SampleSearPointP2VanDiemen2017.csv"
df = pd.read_csv(file_path, skiprows=14, delimiter=",")  # Ajuste conforme necessário

# Função para converter o tempo no formato HH:MM:SS.sss em segundos
def hms_to_seconds(hms):
    try:
        h, m, s = hms.split(':')
        h, m = int(h), int(m)
        s = float(s)
        return h * 3600 + m * 60 + s
    except:
        return None

# Converter a coluna 'Session Time'
df['Session Time'] = df['Session Time'].apply(hms_to_seconds)

# Somar todos os segundos
total_seconds = df['Session Time'].sum()

# Converter o tempo total em horas, minutos e segundos
hours = int(total_seconds // 3600)
minutes = int((total_seconds % 3600) // 60)
seconds = total_seconds % 60

print(f"Tempo Total: {hours} horas, {minutes} minutos e {seconds:.2f} segundos")
```
