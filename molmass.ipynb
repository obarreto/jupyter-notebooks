```python
from molmass import Formula
```

# ETANOL

```python
formula = Formula ('C2H5OH')
formula.mass #g/mol
```

```python
##O Molmass aceita diferentes entrada para um mesmo composto. Vamos passar outra
## string com abreviação do radical etil na orgânica, Et, para obter a massa molar do etanol:

formula = Formula('EtOH')
formula.mass
```

#### Para obter informações da composição elementar do composto. podemos utilizar o método 'composition'

```python
formula.composition()
```
