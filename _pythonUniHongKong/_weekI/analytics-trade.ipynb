```python
# Shift move uma coluna inteira para cima
df['Price1'] = df['Close'].shift(-1)
```

```python
df['Price1']
```

```python
df['PriceDiff'] = df['Price1'] - df['Close']
```

```python
# PriceDiff = (Close price of tomorrow - Close price of today)
df['PriceDiff']
```

```python
df.head()
```

```python
df['Return'] = df['PriceDiff']/df['Close']
```

```python
df.head()
```

```python
df['Close']
```

```python
#Create "Direction" column. Se o preço for positivo atribuíra 1 caso contrário -1
df['Direction'] = [1 if df.loc[ei, 'PriceDiff'] > 0 else -1
                   for ei in df.index]
```

```python
df['Direction']
```

```python
#Create "Moving average" column
df['Average3'] = (df['Close'] + df['Close'].shift(1) + df['Close'].shift(2))/3
```

```python
df['Average3']
```

```python
# Shift (1) é deslocar a coluna inteira para baixo em uma linha
df['Close'].shift(1)
```

```python
# Shift (2) é deslocar a coluna inteira para baixo em duas linha
df['Close'].shift(2)
```

```python
df.head()
```

```python
#Calculate moving average using .rolling()
df['MA40'] = df['Close'].rolling(40).mean()
df['MA200'] = df['Close'].rolling(200).mean()
```

```python
df['Close'].plot(legend=True)
df['MA40'].plot(legend=True)
df['MA200'].plot(legend=True)
```

```python
#Strategy - Fast signal and slow signal
df['MA10'] = df['Close'].rolling(10).mean() #fast signal
df['MA50'] = df['Close'].rolling(50).mean() #slow signal
```

```python
df['Close'].plot(legend=True)
df['MA10'].plot(legend=True)
df['MA50'].plot(legend=True)
```

```python
#Shares is equal to 1 if MA10 is larger than MA50 otherwise it is zero
df['Shares'] = [1 if df.loc[ei, 'MA10'] > df.loc[ei, 'MA50'] else 0
               for ei in df.index]
```

```python
#Daily profit
df['Close1'] = df['Close'].shift(-1)
```

```python
df.iloc[500:505,:]
```

```python
"""
It can be positive or negative.
If negative, we lose money on that day.
If shares is equal to 0,
it means we do not have stock at hand,
the profit is equal to zero
"""
df['Profit'] = [df.loc[ei,'Close1'] - df.loc[ei, 'Close']
               if df.loc[ei,'Shares']==1
               else 0 for ei in df.index]
```

```python
df['Profit'] = [df.loc[ei, 'Close1'] - df.loc[ei, 'Close']
               if df.loc[ei, 'Shares'] == 1
               else 0 for ei in df.index]
df['Profit'].plot()
plt.axhline(y=0, color = 'red')
```

```python
#Cumulative wealth
df['wealth']=df['Profit'].cumsum()
df.tail()
```

```python
print("Total money you win is", df.loc[df.index[-2],'wealth'])
print("Total money you spent is", df.loc[df.index[0],'Close'])
```
