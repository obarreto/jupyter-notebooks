```python
import heapq

# Grafo com custos
graph = {
    'A': {'B': 1, 'C':3},
    'B': {'D' : 1},
    'C': {'E' : 1},
    'D': {'F': 2},
    'E': {'F': 1},
    'F':{}
}

# Heuristica esti,ada (h(n) para A* -> a_star
heuristics = {
    'A': 5,
    'B': 4,
    'C': 2,
    'D': 2,
    'E': 1,
    'F':0
}

def a_star(start, goal):
    frontier = [(heuristics[start], 0, start, [])] #(f, g, node, path)
    visited = set()
    
    while frontier:
        f, g, node, path = heapq.heappop(frontier)

        if node in visited:
            continue
            visited.add(node)

            path += [node]

            if node == goal:
                return path, g

            for neighbor, cost in graph[node].items():
                if neighbor not in visited:
                    g_new = g + cost
                    f_new = g_new + heuristics[neighbor]
                    heapq.heappush(frontier, (f_new, g_new, neighbor, path))
                    
                    #heapq.heappush(frontier, (g + cost + heuristics[neighbnnor], g + cost, neighbor, path))
            return None #, float('inf')

def uniform_cost(start, goal):
    frontier = [(0, start, [])] #(g, node, path)
    visited = set()

    while frontier:
        g, node, path = heapq.heappop(frontier)

        if node in visited:
            continue
        visited.add(node)

        path = path + [node]

        if node == goal:
            return path, g

        for neighbor, cost in graph[node].items():
            if neighbor not in visited:
                heapq.heappush(frontier, (g + cost, neighbor, path))
    return None #, float('inf')

# Executar testes

result_a_star = a_star('A', 'F')
if result_a_star:
    path, cost = result_a_star
    print("A* Search:")
    print("Caminho:", path, "| Custo: ", cost)
else:
    print("A*, Search: Nenhum caminho encontrado.")

result_uniform = uniform_cost('A', 'F')
if result_uniform:
    path, cost = result_uniform
    print("\nCusto Uniforme:")
    print("Caminho:", path, "| Custo:", cost)
else:
    print("Custo Uniforme: Nenhum a caminho encontrado.")
```
