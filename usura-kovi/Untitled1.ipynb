```python
import matplotlib.pyplot as plt

# Calcula a frequência dos valores na coluna 'Description'
frequencia = df['Description'].value_counts()

# Cria um gráfico de barras
plt.figure(figsize=(10, 6))
frequencia.plot(kind='bar')
plt.title('Frequência das Descrições')
plt.xlabel('Descrição')
plt.ylabel('Frequência')
plt.xticks(rotation=90)  # Rotaciona os rótulos no eixo x para melhor visualização
plt.show()
```

```python
# Remover o símbolo "R$" e converter os valores para float
df['Pay'] = df['Pay'].replace('[R$ ]', '', regex=True).str.replace('.', '').str.replace(',', '.').astype(float)

# Somar os valores da coluna 'Pay'
total_pay = df['Pay'].sum()

print("Total Pay:", total_pay)
```

```python
# Remover o símbolo "R$" e converter os valores para float na coluna 'Multa'
# Primeiro, removemos os espaços em branco extras
df['Multa'] = df['Multa'].str.strip()

# Agora, removemos o símbolo "R$" e os pontos, e substituímos a vírgula por ponto
df['Multa'] = df['Multa'].replace('[R$ ]', '', regex=True).str.replace('.', '').str.replace(',', '.')

# Em seguida, substituímos valores vazios ou não numéricos por NaN (Not a Number)
df['Multa'] = pd.to_numeric(df['Multa'], errors='coerce')

# Agora, podemos somar os valores, ignorando os NaN
total_multa = df['Multa'].sum()

print("Total Multa:", total_multa)
```

```python
# Primeiro removemos os espaços em branco extras
df['Usura'] = df['Usura'].str.strip()

# Remover o símbolo "R$" e converter os valores para float
df['Usura'] = df['Usura'].replace('[R$ ]', '', regex=True).str.replace('.', '').str.replace(',', '.')

# Em seguida, substituímos valores vazios ou não numéricos por NaN (Not a Number)
df['Usura'] = pd.to_numeric(df['Usura'], errors='coerce')

# Somamos os valores da coluna 'Usura'
total_usura = df['Usura'].sum()

print("Total Usura:", total_usura)
```
